```python
import os
import xml.etree.ElementTree as ET
from PIL import Image
import shutil

# 設定圖片和標註檔案的資料夾
image_folder = "/content/drive/MyDrive/縮減版＿資料科學dog breeds彙整/Dog_breeds_that_all_three_datasets_have/yorkshire terrier dog/Low_resolution"
annotation_folder = "/content/drive/MyDrive/縮減版＿資料科學dog breeds彙整/Dog_breeds_that_all_three_datasets_have/yorkshire terrier dog/Low_resolution/340-n000120-Yorkshire_terrier"  # 標註檔案資料夾
output_images_folder = "/content/drive/MyDrive/yolo_dog_breeds/dataset/images"
output_labels_folder = "/content/drive/MyDrive/yolo_dog_breeds/dataset/labels"

# 如果輸出資料夾不存在，就建立
os.makedirs(output_images_folder, exist_ok=True)
os.makedirs(output_labels_folder, exist_ok=True)

# 類別對應 (範例：`affenpinscher` 對應到 class_id 0)
class_mapping = {"Yorkshire_terrier": 18}  # 可根據需求添加多類別

# 函數：將 XML 格式轉換為 YOLO 格式
def convert_to_yolo_format(annotation_content, output_txt_file, image_file):
    try:
        # 使用 ElementTree 解析 XML 字串
        root = ET.fromstring(annotation_content)

        # 取得圖片的寬高
        img_width = int(root.find("size/width").text)
        img_height = int(root.find("size/height").text)

        # 建立 YOLO 格式的標註檔案
        with open(output_txt_file, "w") as txt_file:
            for obj in root.findall("object"):
                # 取得類別名稱並轉換為類別索引
                class_name = obj.find("name").text
                if class_name not in class_mapping:
                    print(f"未知類別：{class_name}，已跳過")
                    continue
                class_id = class_mapping[class_name]

                # 取得標註框資訊
                bndbox = obj.find("bndbox")
                xmin = float(bndbox.find("xmin").text)
                ymin = float(bndbox.find("ymin").text)
                xmax = float(bndbox.find("xmax").text)
                ymax = float(bndbox.find("ymax").text)

                # 計算中心點、寬高，並歸一化
                x_center = ((xmin + xmax) / 2) / img_width
                y_center = ((ymin + ymax) / 2) / img_height
                width = (xmax - xmin) / img_width
                height = (ymax - ymin) / img_height

                # 寫入 YOLO 格式
                txt_file.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
    except ET.ParseError as e:
        print(f"解析 XML 時發生錯誤：{e}")
    except Exception as e:
        print(f"其他錯誤：{e}")

# 遍歷圖片資料夾並處理
for image_file in os.listdir(image_folder):
    if image_file.lower().endswith(('.jpg', '.png')):  # 確保處理圖片檔案
        image_name = os.path.splitext(image_file)[0]

        # 查找與圖片名稱匹配的標註檔案
        matched_annotation_file = None
        for annotation_file in os.listdir(annotation_folder):
            if os.path.splitext(annotation_file)[0] == image_name:  # 名稱匹配即可
                matched_annotation_file = os.path.join(annotation_folder, annotation_file)
                break

        # 如果找到對應的標註檔案
        if matched_annotation_file:
            try:
                with open(matched_annotation_file, "r", encoding="utf-8") as f:
                    content = f.read().strip()
                    if "<annotation>" in content:  # 檢查是否包含 XML 結構的核心標籤
                        # 轉換標註為 YOLO 格式
                        output_txt_file = os.path.join(output_labels_folder, image_name + ".txt")
                        convert_to_yolo_format(content, output_txt_file, os.path.join(image_folder, image_file))
                        print(f"已處理標註檔案：{matched_annotation_file}")

                        # 複製圖片到輸出資料夾
                        shutil.copy(os.path.join(image_folder, image_file), os.path.join(output_images_folder, image_file))
                    else:
                        print(f"檔案內容非 XML 格式，已跳過：{matched_annotation_file}")
            except Exception as e:
                print(f"無法處理檔案：{matched_annotation_file}，錯誤：{e}")
        else:
            print(f"未找到對應的標註檔案：{image_file}")

print("轉換完成，資料已存放於指定的 /images 和 /labels 資料夾中。")
```

```python
import os
import shutil
import xml.etree.ElementTree as ET

# 資料夾路徑
images_dir = "/content/drive/MyDrive/縮減版＿資料科學dog breeds彙整/Dog_breeds_that_all_three_datasets_have/affenpinscher dog/Low_resolution"  # 圖片來源資料夾
xml_dir = "/content/drive/MyDrive/縮減版＿資料科學dog breeds彙整/Dog_breeds_that_all_three_datasets_have/affenpinscher dog/Low_resolution/200-n000012-affenpinscher"  # 標註來源資料夾
output_images_dir = "/content/drive/MyDrive/yolo_dog_breeds/dataset/images"  # 輸出圖片資料夾
output_labels_dir = "/content/drive/MyDrive/yolo_dog_breeds/dataset/labels"  # 輸出標註資料夾


# 創建輸出資料夾
os.makedirs(output_images_dir, exist_ok=True)
os.makedirs(output_labels_dir, exist_ok=True)

# 類別對應字典（修改為你的實際類別）
classes = {"affenpinscher": 0 }  # 修改為你的類別名稱

for xml_file in os.listdir(xml_dir):
    if not xml_file.endswith(".xml"):
        continue

    xml_path = os.path.join(xml_dir, xml_file)
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # 圖片資訊
    image_file = root.find("filename").text
    image_path = os.path.join(images_dir, image_file)
    if not os.path.exists(image_path):
        print(f"未找到對應的圖片檔案：{image_file}")
        continue

    # 圖片尺寸
    size = root.find("size")
    image_width = int(size.find("width").text)
    image_height = int(size.find("height").text)

    yolo_annotations = []

    # 遍歷所有物件
    for obj in root.findall("object"):
        class_name = obj.find("name").text
        if class_name not in classes:
            print(f"未知類別：{class_name}，已跳過")
            continue

        class_id = classes[class_name]  # 從字典中獲取類別 ID

        # 處理 headbndbox
        headbndbox = obj.find("headbndbox")
        if headbndbox is not None:
            xmin = float(headbndbox.find("xmin").text)
            ymin = float(headbndbox.find("ymin").text)
            xmax = float(headbndbox.find("xmax").text)
            ymax = float(headbndbox.find("ymax").text)

            # 計算 YOLO 格式的座標
            x_center = ((xmin + xmax) / 2) / image_width
            y_center = ((ymin + ymax) / 2) / image_height
            width = (xmax - xmin) / image_width
            height = (ymax - ymin) / image_height
            yolo_annotations.append(f"{class_id} {x_center} {y_center} {width} {height}")

        # 處理 bodybndbox
        bodybndbox = obj.find("bodybndbox")
        if bodybndbox is not None:
            xmin = float(bodybndbox.find("xmin").text)
            ymin = float(bodybndbox.find("ymin").text)
            xmax = float(bodybndbox.find("xmax").text)
            ymax = float(bodybndbox.find("ymax").text)

            # 計算 YOLO 格式的座標
            x_center = ((xmin + xmax) / 2) / image_width
            y_center = ((ymin + ymax) / 2) / image_height
            width = (xmax - xmin) / image_width
            height = (ymax - ymin) / image_height
            yolo_annotations.append(f"{class_id} {x_center} {y_center} {width} {height}")

    # 將 YOLO 標註存檔
    label_file = os.path.splitext(image_file)[0] + ".txt"
    label_path = os.path.join(output_labels_dir, label_file)
    with open(label_path, "w") as f:
        f.write("\n".join(yolo_annotations))

    # 複製圖片到輸出資料夾
    output_image_path = os.path.join(output_images_dir, image_file)
    if not os.path.exists(output_image_path):
        os.system(f"cp '{image_path}' '{output_image_path}'")

print("轉換完成！資料已存放於指定的 /images 與 /labels 資料夾中。")
```
